```python
import json
# import plt
from matplotlib import pyplot as plt

json_path = "results.json"
with open(json_path) as f:
    data = json.load(f)

# find max
find_index = 10
max_value = 0
min_value = 1
max_version = 0
for i, cur_result_list in enumerate(data):
    actual_index = find_index - 2
    cur_value = cur_result_list[actual_index]
    if cur_value > max_value:
        max_value = cur_value
        max_version = i + 1
    if cur_value < min_value:
        min_value = cur_value

print(f"max: {max_value}, min: {min_value}", f"max_version: {max_version}")


x = [f"1e-{i}" for i in range(2, 14)]

for i, cur_result_list in enumerate(data):
    # draw scatter
    version = i + 1
    plt.scatter(x, cur_result_list, label=f"v{version}")

```

```python
# len(data)

result_list = []

prev_value = 0.99
for find_idx in range(2, 14):
    actual_idx = find_idx - 2
    # find max but lower than prev
    cur_max = 0.0
    cur_idx = 0

    for i, cur_json_res_list in enumerate(data):
        cur_value = cur_json_res_list[actual_idx]
        if cur_value < prev_value and cur_value > cur_max:
            cur_max = cur_value
            cur_idx = i + 1

    prev_value = cur_max

    result_list.append({
        "idx": find_idx,
        "value": cur_max,
        "version": cur_idx
    })

print(result_list)

for res in result_list:
    print(res["idx"], res["version"], res["value"])
    # print(f'"1.0_1e-{res["idx"]}_V{res["version"]}",')



```
